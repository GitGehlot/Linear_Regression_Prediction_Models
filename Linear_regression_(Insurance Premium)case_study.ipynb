```python
#Upload the data using numpy and pandas Lib.

import numpy as np
import pandas as pd
ins_data=pd.read_csv("/content/insurance (2)_linearregression.csv")
```

```python
ins_data.info()  #know your data wisely and feature of each.
ins_data.describe()
ins_data.head()
```

```python
# clean the data and change categoricalvalue to numerical
ins_data['smoker']=ins_data['smoker'].apply({'no':0,'yes':1}.get)
```

```python
ins_data['sex']=ins_data['sex'].apply({'male':0,'female':1}.get)
```

```python
ins_data['region']=ins_data['region'].apply({'northwest':1,'northeast':2,'southeast':3,'southwest':4}.get)
ins_data.head()
```

```python
#Divide the data into independent (x)and dependent (Y) variable
x=ins_data[['age','sex','bmi','children','smoker','region']]
```

```python
y=ins_data['charges']
x.head()
```

```python
y.head()
```

```python
##split the data into training and testing models
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape #check the dimensions after train and test conditions applied
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#create a model with Linear regression model
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
```

```python
regression.fit(x_train,y_train)
predictions=regression.predict(x_test)
```

```python
prediction=regression.predict(x_test)
```

```python
print(prediction)
```

```python
##check the performance of a model
regression.score(x,y)
```
