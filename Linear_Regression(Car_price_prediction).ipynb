**Import the Dependencies**

```python
import numpy as np
import pandas as pd
ins_data=pd.read_csv("/content/Car details.csv") #Data collection and processing
```

```python
ins_data.info()  # Getting some information about the data
ins_data.describe() # Getting some statistical measure about the data
ins_data.head() # Printing the first 5 rows of the dataframe
```

```python
ins_data.isnull().sum()  ## Check the number of missing values in each columns
```

```python
#droping unwanted cloumns as a part of cleaning process

ins_data=ins_data.drop(['name','seller_type','owner','max_power','torque'],axis=1)
```

```python
ins_data.head(30)
```

```python
# checking the distribution of categorical data
ins_data.fuel.value_counts()
```

```python
ins_data.transmission.value_counts()
```

**Encoding the categorical data**

```python
# clean the data and change categoricalvalue to numerical
ins_data.fuel=ins_data.fuel.replace({'Petrol':0,'Diesel':1,'LPG':2,'CNG':3})
```

```python
ins_data['transmission']=ins_data['transmission'].apply({'Manual':0,'Automatic':1}.get)
```

```python
#code for drop text(Unit's) in cloumn section with interger value
ins_data['engine']=ins_data['engine'].str.split('CC').str[0]
ins_data['mileage']=ins_data['mileage'].str.split('kmpl').str[0]
ins_data['mileage']=ins_data['mileage'].str.split('km/kg').str[0]
```

```python
#Removing Rows with NaN Values
ins_data = ins_data.dropna()
```

```python
#divide the data into dependent and independent variable X & Y respectively.
x=ins_data.drop(['selling_price'],axis=1)
```

```python
y=ins_data['selling_price']
```

```python
x.head()
```

```python
x.info()
```

```python
ins_data['mileage'] = ins_data['mileage'].astype(float)

```

```python
ins_data['engine'] = ins_data['engine'].astype(float)
```

```python
ins_data.info()
```

**Spliting the data into Train and Test data**


```python
##split the data into training and testing models
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
x_train.shape #check the dimensions after train and test conditions applied

```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

**Model Training**

```python
#create a model with Linear regression model
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

**Model Evaluation**

```python
predictions=regression.predict(x_test)
```

```python
print(predictions)
```

```python
##check the performance of a model
regression.score(x,y)
```
